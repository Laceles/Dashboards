```python
import dash
import dash_bootstrap_components as dbc
import dash_core_components as dcc
import dash_html_components as html

# Criação do aplicativo Dash
app = dash.Dash(__name__, external_stylesheets=[dbc.themes.BOOTSTRAP])

# Layout do aplicativo
app.layout = html.Div(
    children=[
        html.H1("Exemplo de Dropdowns Alinhados"),
        dbc.Row(
            children=[
                dbc.Col(
                    dcc.Dropdown(
                        id='dropdown-1',
                        options=[
                            {'label': 'Opção 1', 'value': 'opcao1'},
                            {'label': 'Opção 2', 'value': 'opcao2'},
                            {'label': 'Opção 3', 'value': 'opcao3'},
                        ],
                        value='opcao1'
                    ),
                    width=3  # Largura do primeiro dropdown
                ),
                dbc.Col(
                    dcc.Dropdown(
                        id='dropdown-2',
                        options=[
                            {'label': 'Opção A', 'value': 'opcaoA'},
                            {'label': 'Opção B', 'value': 'opcaoB'},
                            {'label': 'Opção C', 'value': 'opcaoC'},
                        ],
                        value='opcaoA'
                    ),
                    width=3  # Largura do segundo dropdown
                ),
                # Adicione mais dbc.Col para mais dropdowns
            ],
            justify='center'  # Centralizar os elementos horizontalmente
        ),
        # Outros componentes e gráficos podem ser adicionados aqui
    ]
)

# Execução do aplicativo
if __name__ == "__main__":
    app.run_server(debug=True)
```
